```python
# Daniel Smith
# Student Achievement
# Allows input of student information to determine if they qualify as achieving Dean's List or Honor Roll


# Configuration variables
deans_list_gpa = 3.5
honor_roll_gpa = 3.25
end_sentinel = "ZZZ"
min_gpa = 0.0
max_gpa = 4.0

# Function to get a non-empty string using a given prompt
def get_string(prompt):
    while True:
            string = input(prompt)

            # String can't be blank
            if string == "":
                print("Input cannot be blank.")
            else:
                return string


# Loop to allow multiple students to be processed
while True:
    # Get a valid last name
    last_name = get_string(f"Enter new student's last name (or {end_sentinel} to quit): ")

    # If the user wanted to exit then let's do that now
    if last_name == end_sentinel:
        break

    # Get a valid first name
    first_name = get_string("Enter the student's first name: ")

    # Get the GPA
    gpa = None
    while True:
        try:
            gpa = float(input(f"Enter {first_name} {last_name}'s GPA: "))

            # If the GPA is out of range raise an error about it
            if gpa < min_gpa or gpa > max_gpa:
                print(f"GPAs must be between {min_gpa} and {max_gpa}.")
            else:
                # Otherwise we have a valid GPA
                break

        except ValueError:
            print(f"GPAs must be a decimal or whole number between {min_gpa} and {max_gpa}.")

    # Determine the level of achievement (assuming to only print one or the other)
    if gpa > deans_list_gpa:
        print(f"{first_name} {last_name} made the Dean's List.")
    elif gpa > honor_roll_gpa:
        print(f"{first_name} {last_name} made the Honor Roll.")

    # Print a separating line for clarity
    print("")

print("Exiting.")
```
